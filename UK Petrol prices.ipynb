### Exploratory Data Aanalysis of the Data
##### This is done to have an overview of the data , this process helps the analyst check the integrity of the data and to enable him/her know the extent of work to be done in the cleaning stage.This dataset only has 1 missing value which is an entire row,after proper inspection of teh data,this row is very insignificant to the whole dataset and can be dropped.

```python
df.describe()
```

```python
from pandas_profiling import ProfileReport
profile=ProfileReport(df,title='Exploratory Data Analysis of Test Data', html={'style' : {'full_width':True}})
```

```python
profile
```

```python
df.isnull().sum()
#From the Inspection of the las row,the entry most likely is an invalid row accounting for the null value in the dataset, since its
#just a row with no values in the columns and very insignificant, we can confidently drop the row, this will enable us a valid dataset
```
